```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'YearBuilt': [2000, 1995, 2010, 1980, 2005,2003,2004,],
    'YearSold': [2011, 2012, 2013, 2014, 2016, 2017,2018],
    'SalePrice': [250000, 200000, 180000, 300000, 280000, 260000,240000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create new feature Age
df['Age'] = df['YearSold'] - df['YearBuilt']

# Plot the relationship between Age and SalePrice
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['SalePrice'], alpha=0.6)
plt.title('Relationship between Age and Sale Price')
plt.xlabel('Age of the Movies (years)')
plt.ylabel('Sale Price ($)')
plt.grid(True)
plt.show()
```
